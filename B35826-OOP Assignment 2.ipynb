```python
"""  
Q1: Electricity Billing Optimization
You are analyzing monthly household electricity consumption data for 10 customers in Kampala. Each record contains: customer_name, units_used, connection_type (e.g., 'Bosco', 245, 'Domestic').
Attempt the following tasks:
1.1.
Use zip() and map() to pair customer names with total bills computed by a lambda: • Domestic rate = 820 UGX/unit • Commercial rate = 1000 UGX/unit
1.2.
Use filter() to separate high-consumption customers (units > 500).
1.3.
Using *args, write a function average_bill(*bills) that computes the average.
1.4.
Display results using list comprehension in the format: ['Bosco: UGX 200,900', 'Jane: UGX 400,000', ...]

"""
```

```python

# 1.1. Use zip() and map() to pair customer names with total bilss computed by a lambda: Domestic rate =820UGX/unit, Commercial rate =100 UGX/Unit
customers = [
    ('Bosco', 245, 'Domestic'),
    ('Jane', 600, 'Commercial'),
    ('David', 420, 'Domestic'),
    ('Sarah', 750, 'Commercial'),
    ('Mike', 320, 'Domestic'),
    ('Grace', 550, 'Commercial'),
    ('Peter', 280, 'Domestic'),
    ('Anna',480, 'Domestic'),
    ('John',680, 'Domestic'),
    ('Mary',380, 'Domestic')
]

#using zip() and map() to pair customer names with total bills
domestic_rate =820
commercial_rate =1000

#creating lambda function for bill calculation
calculate_bills =lambda customer:customer[1]*domestic_rate if customer[2] =='Domestic' else customer[1] *commercial_rate

#using map to calculate bills and zip to pair names
customer_bills =list(zip( 
          [customer[0] for customer in customers],
           map(calculate_bills, customers)
))

print("1.1. Customer Bills:", customer_bills)

    
```

```python
#1.2 use filter() to separate high_consumption customers (units>500)

high_consumption =list(filter (lambda customer: customer[1] > 500, customers))
print("1.2 High Consumption Customers:", high_consumption)
```

```python
#1.3. Using *args, write a function average_bill (*bills) that computes the average

def average_bill(*bills):
    return sum(bills)/len(bills) if bills else 0

bill_amounts = [bill for _, bill in customer_bills]
avg_bill = average_bill(*bill_amounts)
print("1.3. Average Bill: UGX", f"{avg_bill:,.0f}")
```

```python
#1.4 Display results using list comparisom
updated_bills = [f"{name}: UGX {bill:,.0f}" for name, bill in customer_bills]

print("1.4 Updated Bills:")

for bill in updated_bills:
    print(" ", bill)
```

```python
""" 

Q2: Market Basket Price Aggregator
Each vendor reports different fruit prices daily:
mango_prices = [2500, 2700, 2600, 2800] orange_prices = [3000, 3200, 3100, 3050] apple_prices = [4500, 4600, 4550, 4700]
Attempt the following tasks:
2.1
Use zip_longest() to handle missing prices.
2.2
Define a lambda function avg = lambda lst: sum(lst)/len(lst) to compute average per fruit.
2.3
Create a generator expression that yields all fruits with average price above 3000.
2.4
Print the generator’s results neatly as: Fruits Above Average Price: Mango, Orange, Apple

"""
```

```python
# Use zip_longest() to handle missing prices

from itertools import zip_longest

#Given data
mango_prices = [2500, 2700,2600, 2800]
orange_prices = [3000,3200, 3100, 2050]
apple_prices = [4500,4600,4550,4700]

print(" Original Prices:")
print("Mango:", mango_prices)
print("Orange:",orange_prices)
print("Apple:", apple_prices)
print ()

#2.1. Use zip_longest() to handle missing prices
all_prices = list(zip_longest(mango_prices, orange_prices,apple_prices,fillvalue=0))
print ("2.1 Combined Prices (Zip_longest):", all_prices)



```

```python
#2.2. Define a lambda function to compute average per fruit

avg = lambda list: sum(list)/len(list)

mango_avg =avg(mango_prices)
orange_avg = avg(orange_prices)
apple_avg = avg(apple_prices)

print("2.2 Average prices:")
print(f" Mango: UGX {mango_avg:.0f}")
print(f" Orange: UGX {orange_avg:.0f}")
print(f" apple: UGX {apple_avg:.0f}")
```

```python
#2.3 Create a generator expression that yields all fruits with average price above 3000

fruit_data =[('mango', mango_avg), ('orange', orange_avg), ('apple', apple_avg)]
above_3000_generator = (fruit for fruit, price in fruit_data if price > 3000)
```

```python
#2.4 Print the genrator's results neatly
print("2.4 Fruits Above Average Price:", end = " ")
fruits_above_3000 = list(above_3000_generator)

print (" ".join(fruits_above_3000))
```

```python
"""   

Q3: District Temperature Tracker
Temperature data (°C):
kampala = [28, 30, 29, 27, 26, 29, 30, 31, 32, 30, 29, 28] gulu = [25, 26, 27, 27, 28, 29, 30, 30, 29, 27, 26, 25] mbarara = [22, 23, 23, 24, 25, 25, 26, 27, 27, 26, 24, 23]
Attempt the following tasks:
3.1. Use map() with lambda to convert all temperatures to Fahrenheit. 
3.2. Use zip() to pair months (Jan–Dec) with Kampala temps.
3.3. Use filter() to identify months with temperature > 30°C. 
3.4. Create a generator to yield month names and temperatures above threshold.
3.5. Print as: Hot months in Kampala: July - 31°C, August - 32°C

"""
```

```python
#3.1. use map() with lambda to convert all temperatures to Fahrenheit.

#Temparature Data 

Kampala = [28, 30,29,27,26,29,30,29,28]
Gulu = [25,26,27,28,29,30,29,27,26,25]
Mbarara = [22, 23, 23,24,25,25,26,27,27,26,24,23]

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# 3.1. Use map() with lambda to convert all temperatures to Fahrenheit
c_to_f = lambda c:(c*9/5) +32
Kampala_f = list(map(c_to_f, Kampala))
Gulu_f = list(map(c_to_f, Gulu))
Mbarara_f = list(map(c_to_f, Mbarara))

print(" 3.1. Temperatures in Fahrenheit")
print("Kampala:",[round(temp,1) for temp in Kampala_f])
print("Gulu:", [round(temp, 1) for temp in Gulu_f])
print("Mbarara:", [round(temp, 1) for temp in Mbarara_f])
                 
```

```python
#3.2 Use zip() to pair (Jan-Dec) with Kampala Temps

Kampala_monthly = list(zip(months, Kampala))
print("\n 3.2. Monthly Kampala Temperatures:", Kampala_monthly)
```

```python
# 3.3 Use filter() to identify months with temperature >30c

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

Kampala = [28, 30, 29, 27, 26, 29, 30, 31, 32, 30, 29, 28]

Kampala_monthly = list(zip(months, Kampala))
hot_months = list(filter(lambda month_temp: month_temp[1] > 30, Kampala_monthly))
print("3.3 Hot Months (>30°C):", hot_months)
```

```python
# 3.4 create a generator to yield month names and tempearatures above threshold
hot_months_generator = (( month, temp) for month, temp in Kampala_monthly if temp> 30)
```

```python
# 3.5 Print as : Hot months in Kampala: July-31C, August-32C

hot_months_list = list(hot_months_generator)
updated_hot_months = [f"{month} - {temp}°C" for month, temp in hot_months_list]
print(" 3.5 Hot Months in Kampala:", " ".join(updated_hot_months))
```

```python
"""   
Q4: School Fees Payment Analyzer
Installment data:
students = ['Alex', 'Grace', 'Sarah', 'Brian'] installments = [ [150000, 200000, 250000], [500000, 0, 200000], [300000, 300000, 300000], [400000, 100000, 0] ]
Attempt the following tasks:
4.1. Write a lambda function that sums valid payments and ignores zeros. 
4.2. Use map() to compute total paid by each student. 
4.3. Combine student names and totals using zip(). 
4.4. Use filter() to find students who have cleared full fees (≥600,000). 
4.5. Create payment_summary(**kwargs) that receives named args like Alex=600000 and prints summaries.

"""
```

```python
#Installement Data

students = [ 'Alex', 'Grace', 'Sarah', 'Brian']
Installments = [[150000, 200000, 250000], [500000, 0, 200000], [300000, 300000,300000],[400000, 100000,0]]
print(" Original Data:")
print("Students:", students)
print("Installments:", Installments)
```

```python
#4.1. Write a lambda function that sums valid payments and ignores Zeros

sum_valid_payments = lambda payments: sum(payment for payment in payments if payment > 0)
print("4.1. Lambda function created:", sum_valid_payments)
```

```python
#4.2 Use map() to compute total paid by each student
total_paid = list(map(sum_valid_payments, Installments))
print("4.3 Student-Total pairs:", total_paid)
```

```python
#4.3 Combine student names and totals using Zip()
student_totals = list(zip(students, total_paid))
print(" 4.3. Student-Total Pairs:", student_totals)
```

```python
#4.4 Use filter() to find students who have cleared full fees (2600,000)
cleared_fees = list(filter(lambda student_total: student_total[1] >= 600000, student_totals))
print("4.4. Students who cleared full fees:", cleared_fees)
```

```python
#4.5 Create payment_summary(**kwargs) that receives named arges
def payment_summary(**kwargs):
    for student, amount in kwargs.items():
        status = "CLEARED" if amount >= 600000 else f"BALANCE: {600000-amount}"
        print(f" {student}: UGX {amount:,} - {status}")

print("4.5 Payment Summary:")
payment_summary(Alex = 600000, Grace = 700000, Sarah =900000, Brian =500000)
```

```python
# Yield Data
districts = [' Bushenyi', 'Mityana', 'Kasese', 'Mbale']
yield_data = [1200, 1500, 900, 1300]

print(" Original Data:")
print("Districts:", districts)
print("Yield_data:",yield_data)
print()
```

```python
#5.1 Use a list comprehension with a lambda to convert yields from kg to tons

kg_to_ton = lambda kg: kg/1000
yield_tons = [kg_to_ton(yield_kg) for yield_kg in yield_data]
print("5.1 Yield in Tons:", [round(ton,2) for ton in yield_tons])
```

```python
#5.2. Write a generator that yields only districts with yield > 1 ton

high_yield_districts = (district for district, ton in zip(districts, yield_tons) if ton >1 )
print("5.2. Districts with yield >1 ton:", list(high_yield_districts))
```

```python
#Use *args to compute average yield across all districts

def average_yield(*yields):
    return sum(yields)/len(yields) if yields else 0
avg_yield = average_yield(*yield_tons)
print("5.3 Average yield:", round(avg_yield, 2), "tons")
```

```python
#5.4. Use **kwargs to estimulate price chnages per district and compute revenue
def calculate_revenue(**district_prices):
    revenues = {}
    for district, price_per_ton in district_prices.items():
        if district in districts:
            index = districts.index(district)
            revenue =yield_tons[index] * price_per_ton 
            revenues[district] = revenue
    return revenues
#stimulate different prices per district (UGX per ton)
revenues = calculate_revenue(Bushenyi = 5000000, Mityana = 4500000, Kasese =5000000, Mbale = 4800000)
print("5.4 District Revenues:", revenues)
```

```python
"""   
Q6: Web Data Aggregation
Given that sites = ['https://ucu.ac.ug', 'https://harba.ug', 'https://www.bou.or.ug']
Attempt the following tasks:
6.1. Write a function that takes *urls and uses requests.get() to return response codes. 
6.2. Use a list comprehension to print only URLs with status code 200. 
6.3. Store results in a dictionary using a dictionary comprehension. 
6.4. Create a generator expression that yields 'Active Site: <url>' for reachable domains.

"""
```

```python
print (" Web Data Aggregation \n")

sites = ['https://ucu.ac.ug', 'https:harba.ug', 'https://bou.or.ug']

# 6.1 Function that takes *urls and returns response codes (simulated)
def get_response_codes(*urls):
    response_codes = {}
    for url in urls:
        # Simulate different response codes for demonstration
        if 'ucu' in url:
            response_codes[url] = 200
        elif 'harba' in url:
            response_codes[url] = 404
        else:
            response_codes[url] = 200
        print(f"Checked {url}: Status {response_codes[url]}")
    return response_codes

print("6.1 Getting response codes...")
response_data = get_response_codes(*sites)
print()
```

```python
#6.2 list comprehension to print only URLs with status code 200

print ("6.2 URLs with status Code 200")
successful_urls = [url for url, code in response_data.items() if code ==200]
for url in successful_urls:
    print (f" Correct {url}")
print()
```

```python
# 6.3 Store results in a dictionary using dictionary comprehension

print("6.3 Results Dictionary:")

results_dict ={url:code for url, code in response_data.items()}
for url, status in results_dict.items():
    status_icon = "Correct" if status == 200 else " Not Correct"
    print( f" {status_icon} {url}: {status}")
print()
```

```python
# 6.4 Generator expression that yields 'Active Site: <url>' for reacheable domains

print("6.4 Active Sites from Generator:")
active_generator = (f'Active site: {url}' for url in successful_urls)
for message in active_generator:
    print(f" {message}")

print("\n=== All tasks completed successfully!===")
```
